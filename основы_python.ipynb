#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(input_string):
    vowel_chars = 'aeiou'
    vowel_count = 0
    
    lowercase_text = input_string.lower()
    
    for character in lowercase_text:
        if character in vowel_chars:
            vowel_count += 1
            
    return vowel_count
```

```python
# Тесты
def test_count_vowels_simple_word():
    result = count_vowels("hello")
    assert result == 2

def test_count_vowels_uppercase():
    result = count_vowels("WORLD")
    assert result == 1

def test_count_vowels_all_vowels():
    result = count_vowels("aeiou")
    assert result == 5

def test_count_vowels_no_vowels():
    result = count_vowels("bcdfg")
    assert result == 0

def test_count_vowels_empty_string():
    result = count_vowels("")
    assert result == 0

def test_count_vowels_mixed_case():
    result = count_vowels("AeIoU")
    assert result == 5

def test_count_vowels_with_spaces():
    result = count_vowels("hello world")
    assert result == 3

def test_count_vowels_special_chars():
    result = count_vowels("hello! world?")
    assert result == 3

# Демонстрация работы
if __name__ == "__main__":
    # Примеры использования
    test_strings = [
        "hello",
        "WORLD", 
        "aeiou",
        "bcdfg",
        "",
        "AeIoU",
        "hello world",
        "Programming is fun!"
    ]
    
    print("Подсчет гласных в строках:")
    print("=" * 30)
    
    for test_str in test_strings:
        vowel_count = count_vowels(test_str)
        print(f"'{test_str}' -> {vowel_count} гласных")
    
    # Запуск тестов
    print("\n" + "=" * 30)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def has_unique_chars(input_string):
    char_set = set()
    for char in input_string:
        if char in char_set:
            return False
        char_set.add(char)
    return True
```

```python
# Тесты
def test_unique_chars_all_unique():
    result = has_unique_chars("abcde")
    assert result == True

def test_unique_chars_duplicates():
    result = has_unique_chars("hello")
    assert result == False

def test_unique_chars_empty():
    result = has_unique_chars("")
    assert result == True

def test_unique_chars_single():
    result = has_unique_chars("a")
    assert result == True

def test_unique_chars_case_sensitive():
    result = has_unique_chars("AaBb")
    assert result == True

def test_unique_chars_special():
    result = has_unique_chars("!@#$%")
    assert result == True

def test_unique_chars_spaces():
    result = has_unique_chars("a bc")
    assert result == True

def test_unique_chars_duplicate_spaces():
    result = has_unique_chars("a  b")
    assert result == False

def test_unique_chars_long_duplicate():
    result = has_unique_chars("a" * 100)
    assert result == False

def test_unique_chars_mixed_whitespace():
    result = has_unique_chars(" \t")
    assert result == True

def test_unique_chars_numbers():
    result = has_unique_chars("12345")
    assert result == True

def test_unique_chars_duplicate_numbers():
    result = has_unique_chars("1123")
    assert result == False

# Демонстрация работы
if __name__ == "__main__":
    test_strings = [
        "abcde",
        "hello",
        "",
        "a",
        "AaBb", 
        "!@#$%",
        "a bc",
        "a  b",
        "12345",
        "1123"
    ]
    
    print("Проверка уникальности символов:")
    print("=" * 40)
    
    for test_str in test_strings:
        is_unique = has_unique_chars(test_str)
        status = "уникальны" if is_unique else "есть дубликаты"
        print(f"'{test_str}' -> {status}")
    
    # Запуск тестов
    print("\n" + "=" * 40)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 0:
        raise ValueError("Число должно быть положительным")
    
    if n < 10:
        return 0
    
    steps = 0
    current = n
    
    while current >= 10:
        product = 1
        for digit in str(current):
            product *= int(digit)
        
        current = product
        steps += 1
    
    return steps
```

```python
# Тесты
def test_magic_single_digit():
    assert magic(5) == 0
    assert magic(9) == 0
    assert magic(0) == 0

def test_magic_two_digits():
    assert magic(10) == 1
    assert magic(25) == 2
    assert magic(39) == 3

def test_magic_three_digits():
    assert magic(100) == 1
    assert magic(999) == 4
    assert magic(123) == 1

def test_magic_large_numbers():
    assert magic(277777788888899) == 11
    assert magic(3777888999) == 10
    assert magic(1111111111) == 1

def test_magic_edge_cases():
    assert magic(1) == 0
    assert magic(9) == 0
    assert magic(10) == 1
    assert magic(99) == 2

def test_magic_known_sequences():
    assert magic(77) == 4
    assert magic(679) == 5
    assert magic(688) == 7

def test_magic_negative_input():
    try:
        magic(-5)
        assert False, "Ожидалось ValueError"
    except ValueError:
        pass

# Демонстрация работы
if __name__ == "__main__":
    test_cases = [
        5, 10, 25, 39, 77, 99, 100, 123, 679, 999, 2777
    ]
    
    print("Количество шагов для получения одной цифры:")
    print("=" * 50)
    print(f"{'Число':<12} {'Шаги':<8} {'Процесс':<30}")
    print("-" * 50)
    
    for num in test_cases:
        steps = magic(num)
        
        # Показываем процесс
        process = []
        current = num
        while current >= 10:
            digits = [int(d) for d in str(current)]
            product = 1
            for d in digits:
                product *= d
            process.append(f"{current}→{product}")
            current = product
        process_str = " → ".join(process)
        
        print(f"{num:<12} {steps:<8} {process_str}")
    
    # Запуск тестов
    print("\n" + "=" * 50)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")

    if len(pred) == 0:
        return 0

    squared_errors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return sum(squared_errors) / len(squared_errors)
```

```python
# Тесты
def test_mse_equal_vectors():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    assert mse([0, 0, 0], [0, 0, 0]) == 0

def test_mse_different_vectors():
    assert mse([1, 2, 3], [4, 5, 6]) == 9
    assert mse([1, 1], [3, 3]) == 4

def test_mse_negative_numbers():
    assert mse([-1, -2], [1, 2]) == 10
    assert mse([-5, 5], [5, -5]) == 100

def test_mse_decimal_numbers():
    assert abs(mse([1.5, 2.5], [1.0, 2.0]) - 0.25) < 1e-10
    assert abs(mse([0.1, 0.2], [0.3, 0.4]) - 0.04) < 1e-10

def test_mse_single_element():
    assert mse([5], [3]) == 4
    assert mse([10], [10]) == 0

def test_mse_empty_vectors():
    assert mse([], []) == 0

def test_mse_different_lengths():
    try:
        mse([1, 2], [1, 2, 3])
        assert False, "Ожидалось ValueError"
    except ValueError:
        pass
    
    try:
        mse([1], [])
        assert False, "Ожидалось ValueError"
    except ValueError:
        pass

def test_mse_large_vectors():
    pred = [i for i in range(100)]
    true = [i + 1 for i in range(100)]
    result = mse(pred, true)
    assert result == 1.0

# Демонстрация работы
if __name__ == '__main__':
    test_cases = [
        ([1, 2, 3], [1, 2, 3]),
        ([1, 2, 3], [4, 5, 6]),
        ([1.5, 2.5], [1.0, 2.0]),
        ([5], [3]),
        ([], [])
    ]
    
    print("Среднеквадратичная ошибка (MSE):")
    print("=" * 60)
    print(f"{'Предсказания':<20} {'Истинные значения':<20} {'MSE':<10}")
    print("-" * 60)
    
    for pred, true in test_cases:
        if len(pred) == len(true):
            error = mse(pred, true)
            print(f"{str(pred):<20} {str(true):<20} {error:<10.4f}")
        else:
            print(f"{str(pred):<20} {str(true):<20} {'ERROR':<10}")
    
    # Запуск тестов
    print("\n" + "=" * 60)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    if n == 1:
        return "(1)"

    factors = {}
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1

    if n > 1:
        factors[n] = factors.get(n, 0) + 1

    result = ""
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{power})"

    return result
```

```python
# Тесты
def test_prime_factors_one():
    assert prime_factors(1) == "(1)"

def test_prime_factors_prime_numbers():
    assert prime_factors(2) == "(2)"
    assert prime_factors(3) == "(3)"
    assert prime_factors(17) == "(17)"
    assert prime_factors(97) == "(97)"

def test_prime_factors_prime_powers():
    assert prime_factors(4) == "(2**2)"
    assert prime_factors(8) == "(2**3)"
    assert prime_factors(9) == "(3**2)"
    assert prime_factors(16) == "(2**4)"
    assert prime_factors(25) == "(5**2)"

def test_prime_factors_composite_numbers():
    assert prime_factors(6) == "(2)(3)"
    assert prime_factors(12) == "(2**2)(3)"
    assert prime_factors(15) == "(3)(5)"
    assert prime_factors(30) == "(2)(3)(5)"

def test_prime_factors_large_numbers():
    assert prime_factors(100) == "(2**2)(5**2)"
    assert prime_factors(12345) == "(3)(5)(823)"
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"

def test_prime_factors_edge_cases():
    assert prime_factors(1) == "(1)"
    assert prime_factors(2) == "(2)"
    assert prime_factors(3) == "(3)"
    assert prime_factors(4) == "(2**2)"

# Демонстрация работы
if __name__ == '__main__':
    test_numbers = [
        1, 2, 3, 4, 6, 8, 9, 12, 15, 16, 
        25, 30, 100, 123, 86240, 97, 1000
    ]
    
    print("Разложение чисел на простые множители:")
    print("=" * 50)
    print(f"{'Число':<8} {'Разложение':<30}")
    print("-" * 50)
    
    for num in test_numbers:
        factors = prime_factors(num)
        print(f"{num:<8} {factors:<30}")
    
    # Запуск тестов
    print("\n" + "=" * 50)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"

    k = 0
    total = 0

    while total < number:
        k += 1
        total += k * k

        if total == number:
            return k

    return "It is impossible"
```

```python
# Тесты
def test_pyramid_small_numbers():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3

def test_pyramid_impossible_cases():
    assert pyramid(2) == "It is impossible"
    assert pyramid(3) == "It is impossible"
    assert pyramid(4) == "It is impossible"
    assert pyramid(6) == "It is impossible"
    assert pyramid(15) == "It is impossible"
    assert pyramid(7) == "It is impossible"

def test_pyramid_larger_numbers():
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(91) == 6
    assert pyramid(140) == 7

def test_pyramid_edge_cases():
    assert pyramid(0) == "It is impossible"
    assert pyramid(-5) == "It is impossible"
    assert pyramid(-100) == "It is impossible"

def test_pyramid_known_sequence():
    pyramidal_numbers = [1, 5, 14, 30, 55, 91, 140, 204, 285, 385]
    for i, num in enumerate(pyramidal_numbers, 1):
        assert pyramid(num) == i

def test_pyramid_large_impossible():
    assert pyramid(100) == "It is impossible"
    assert pyramid(200) == "It is impossible"
    assert pyramid(300) == "It is impossible"

# Демонстрация работы
if __name__ == '__main__':
    test_numbers = [
        1, 2, 5, 6, 14, 15, 30, 31, 55, 56, 
        91, 92, 140, 0, -5, 100, 204, 385
    ]
    
    print("Проверка пирамидальных чисел:")
    print("=" * 45)
    print(f"{'Число':<8} {'Результат':<20} {'Объяснение':<30}")
    print("-" * 45)
    
    for num in test_numbers:
        result = pyramid(num)
        if result == "It is impossible":
            explanation = "Не является пирамидальным"
        else:
            explanation = f"1² + 2² + ... + {result}² = {num}"
        print(f"{num:<8} {str(result):<20} {explanation:<30}")
    
    # Запуск тестов
    print("\n" + "=" * 45)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number):
    digits = list(map(int, str(number)))
    n = len(digits)

    if n <= 2:
        return True

    if n % 2 == 0:
        mid_start = n // 2 - 1
        mid_end = n // 2
        left = digits[:mid_start]
        right = digits[mid_end + 1:]
    else:
        mid = n // 2
        left = digits[:mid]
        right = digits[mid + 1:]

    return sum(left) == sum(right)
```

```python
# Тесты
def test_is_balanced_single_digit():
    assert is_balanced(5) == True
    assert is_balanced(0) == True
    assert is_balanced(9) == True

def test_is_balanced_two_digits():
    assert is_balanced(12) == True
    assert is_balanced(99) == True
    assert is_balanced(10) == True
    assert is_balanced(55) == True

def test_is_balanced_odd_balanced():
    assert is_balanced(121) == True
    assert is_balanced(10301) == True
    assert is_balanced(12321) == True
    assert is_balanced(10001) == True

def test_is_balanced_odd_unbalanced():
    assert is_balanced(123) == False
    assert is_balanced(10234) == False
    assert is_balanced(12345) == False
    assert is_balanced(10012) == False

def test_is_balanced_even_balanced():
    assert is_balanced(1221) == True
    assert is_balanced(1001) == True
    assert is_balanced(1331) == True
    assert is_balanced(2332) == True

def test_is_balanced_even_unbalanced():
    assert is_balanced(1234) == False
    assert is_balanced(1002) == False
    assert is_balanced(1223) == False
    assert is_balanced(2333) == False

def test_is_balanced_with_zeros():
    assert is_balanced(100001) == True
    assert is_balanced(100012) == False
    assert is_balanced(1001001) == True
    assert is_balanced(1002001) == False

def test_is_balanced_large_numbers():
    assert is_balanced(1234554321) == True
    assert is_balanced(12345654321) == True
    assert is_balanced(123456789) == False

# Демонстрация работы
if __name__ == '__main__':
    test_numbers = [
        5, 12, 99, 121, 123, 1221, 1234, 
        1001, 10301, 100001, 100012,
        12321, 1234554321, 123456789
    ]
    
    print("Проверка сбалансированных чисел:")
    print("=" * 55)
    print(f"{'Число':<15} {'Результат':<12} {'Объяснение':<30}")
    print("-" * 55)
    
    for num in test_numbers:
        result = is_balanced(num)
        digits = list(map(int, str(num)))
        n = len(digits)
        
        if n <= 2:
            explanation = "1-2 цифры - всегда сбалансировано"
        elif n % 2 == 0:
            mid_start = n // 2 - 1
            mid_end = n // 2
            left_sum = sum(digits[:mid_start])
            right_sum = sum(digits[mid_end + 1:])
            explanation = f"sum({digits[:mid_start]}) = {left_sum}, sum({digits[mid_end + 1:]}) = {right_sum}"
        else:
            mid = n // 2
            left_sum = sum(digits[:mid])
            right_sum = sum(digits[mid + 1:])
            explanation = f"sum({digits[:mid]}) = {left_sum}, sum({digits[mid + 1:]}) = {right_sum}"
        
        status = "Сбалансировано" if result else "Не сбалансировано"
        print(f"{num:<15} {status:<12} {explanation:<30}")
    
    # Запуск тестов
    print("\n" + "=" * 55)
    print("Запуск тестов...")
    
    import pytest
    pytest.main([__file__, "-v"])
```
